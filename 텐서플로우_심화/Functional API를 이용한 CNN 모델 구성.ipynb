## CNN 모델을 이용한 CIFAR-10 데이터 셋 분류

```python
from tensorflow.keras.datasets import cifar10
```

* Step 1. Input tensor 와 Target tensor 준비(훈련데이터)

```python
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()
```

* Step 2. 입력데이터의 전처리 

```python
train_images = train_images/255.
test_images = test_images/255.
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
train_labels.shape, test_labels.shape
```

* Step3. CNN 모델 디자인

```python
# feature_extracion :
#  - conv layer(filter=32, size=(3,3)), pooling layer(max, (2, 2))
#  - conv layer(filter=64, size=(3,3)), pooling layer(max, (2, 2))
#  - conv layer(filter=64, size=(3,3))
# flatten : 
# classification : 
#  - label 데이터를 고려하여 직접 디자인
```

```python
from tensorflow import keras
```

```python
from tensorflow.keras import models, layers
```

```python
# 모델 구성
input = keras.Input(shape=(32, 32, 3))

x = layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu')(input)
x = layers.MaxPooling2D(pool_size=(2,2))(x)
x = layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu')(x)
x = layers.MaxPooling2D(pool_size=(2,2))(x)
x = layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu')(x)

x = layers.Flatten()(x)

x = layers.Dense(64, 'relu')(x)
output = layers.Dense(10, 'softmax')(x)
```

```python
# 모델 객체 생성
model = keras.Model(
    inputs=input,
    outputs=output,
    name='CIFAR_CNN'
)
```

```python
model.summary()
```

* Step4. 학습에 필요한 정보 설정

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

* Step5. 모델에 데이터 연결하여 학습

```python
model.fit(x=train_images, y=train_labels,
          epochs=10,
          batch_size=32,
          validation_split=0.2)
```

* Step6. test 데이터 셋을 이용하여 모델의 성능 평가

```python
test_loss, test_accuracy = model.evaluate(x=test_images, y=test_labels)
```
